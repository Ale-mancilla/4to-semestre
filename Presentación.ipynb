**CONSTRUCCIÓN SUSTENTABLE

- Dentro de la ingenieria civil podemos encontrarnos con la principal problematica de el impacto ambiental y las pocas    alternativas existentes, es importante que los proyectistas se enfoquen en la resolución del problema y busquen soluciones   que sean lo menor posible dañinas para el medio ambiente.

- A este concepto se le conoce como proyecto de desarrollo sustentable que es el conjunto de todas las ideas puestas en ejecución con el objetivo de obtener prosperidad económica basada en la equidad social y el cuidado del medio ambiente.
- Son ideas que abordan las problemáticas del desarrollo sustentable, es decir, que tienen como objetivo la prosperidad económica, la integridad del medio ambiente y la equidad social.
